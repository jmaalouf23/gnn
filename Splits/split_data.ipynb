```python
import os
import sys
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_path='../Data/thermo/SI_data_CombiSolv_Exp_QM_csv.csv'
df=pd.read_csv(data_path)
N=len(df.index)
data=list(range(N))
print(f'There are {N} data points ')
```

# Random splits

```python
"""
Will create splits for chiral_gnn where the test set is the same, but the train and val sets get chosen randomly
"""
num_splits=5
split_dir='thermo/solvation/80_10_20'
os.makedirs(split_dir, exist_ok=True)

train, test= train_test_split(data, test_size=0.2,shuffle=True)
    
for i in range(num_splits):
    
    temp_train, temp_val= train_test_split(train, test_size=0.125,shuffle=True)
    splits=[temp_train,temp_val,test]
    np.save(f'{split_dir}/split_{i}.npy',splits)
```

```python
print(f'Length of Train is {len(temp_train)}')
print(f'Length of Val is {len(temp_val)}')
print(f'Length of Test is {len(test)}')
```
